### 3.3 Skewness and Distribution

###### a positively skewed (or right-skewed) distribution is a type of distribution in which most values are clustered around the left tail of the distribution while the right tail of the distribution is longer.

```python
SalePrice = df['SalePrice']
SalePrice.skew()
```

```python
#plotting histogram to check distribution of SalePrice as its Dependant variable
import seaborn as sns
sns.set(rc={'figure.figsize':(5,4)})
sns.distplot(df['SalePrice'],bins=20);
```

#### CHI Square and PValues

```python
# Swarm Plot
plt.figure(figsize=(12,6))
sns.swarmplot('OverallQual', 'SalePrice',data=df)
```

#### There is a marked increase in the saleprice with an increase in the overall quality. So it is an important predictor for saleprice

```python
# Testing different independant variables with chi square test
from scipy.stats import chi2_contingency
chiSq = df.loc[:,['MSSubClass','LotArea']]
stat, p,dof,expected = chi2_contingency(chiSq)
p
```

#### p is less than 0.05 so reject null hypothesis vch says theres is no relationship
